Suppose you have a list named my_list as given below.

my_list = [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]

Now use bubble sort to sort my_list into ascending order.

===============================================================
Output:

[1, 2, 2, 3, 5, 6, 10, 11, 12, 14, 15, 17, 18, 20, 29]

```python
my_list = [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29] 

for index_1 in range(0, len(my_list)-1): 
    for index_2 in range(0, len(my_list)- index_1 - 1):

        if (my_list[index_2] > my_list[index_2 + 1]):

            temp = my_list[index_2]
            my_list[index_2] = my_list[index_2 + 1]
            my_list[index_2 + 1] = temp
print( my_list)
```

my_list = [4, 2, 3, 1, 6, 5]

The sorted list would be [1, 2, 3, 4, 5, 6] where 4 numbers (4, 1, 6, and 5) have changed their positions. Therefore our sample output here would be 4. Please check if your code is working correctly by changing the above list and using the knowledge given here to verify whether your code gives the correct output for all different lists.

Sample Output for the above list

4

```python
my_list = [4, 2, 3, 1, 6, 5]
def selection_sort(list_a):
    list_b = list_a.copy()
    for index_1 in range(len(list_a)):
        min_index = index_1
        for index_2 in range(index_1 + 1, len(list_a)):
            if list_a[index_2] < list_a[min_index]:
                min_index = index_2
        list_a[index_1], list_a[min_index] = list_a[min_index], list_a[index_1]
    count = 0
    for itaretion in range(len(list_a)):
        if list_a[itaretion] != list_b[itaretion]:
            count = count + 1
    return(count)
print(selection_sort(my_list))
```

TASK 07
Write a python program that takes two lists from the user, merges the two lists, sorts the resulting list, and then finds the median of the elements in the two lists.

=========================================

Sample Input 1

list_one = [1, 2, 1, 4]

list_two = [5, 4, 1]

Sample Output 1

Sorted list = [1, 1, 1, 2, 4, 4, 5]

Median = 2

=========================================

Sample Input 2

list_one = [1, 7, 9, 10]

list_two = [2, 7, 6, 5]

Sample Output 2

Sorted list = [1, 2, 5, 6, 7, 7, 9, 10]

Median = 6.5

```python
list_one = [1, 2, 1, 4]
list_two = [5, 4, 1]
list_a = list_one + list_two
def bubble_sort(list_b):
    for index_1 in range(len(list_b)):
        for index_2 in range(len(list_b) - 1):
            if list_b[index_2] > list_b[index_2 + 1]:
                list_b[index_2], list_b[index_2 + 1] = list_b[index_2 + 1], list_b[index_2]
    return(list_b)
print('Sorted list= ', bubble_sort(list_a))
if len(list_a) % 2 != 0:
    val = len(list_a) // 2
    print('Median', list_a[val])
else:
    val = len(list_a) / 2
    val_2 = val + 1
    print('Median', (list_a[val] + list_a[val_2] / 2))
```
